```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Find the total number of schools
unique_schools = school_data_complete["school_name"].unique()
num_of_schools = len(unique_schools)
num_of_schools
```

```python
#Find the total number of students
total_num_students = school_data_complete["student_name"].count()
total_num_students
```

```python
#Find the total budget
unique_budget = school_data_complete["budget"].unique()
unique_budget_df = pd.DataFrame(unique_budget)
total_budget = unique_budget_df[0].sum()
total_budget
```

```python
#Find the average math score
avg_math_score = school_data_complete["math_score"].mean()
avg_math_score
```

```python
#Find the average reading score
avg_reading_score = school_data_complete["reading_score"].mean()
avg_reading_score
```

```python
#Find the percentage of students with a passing (70 or greater) math score
math_passing_df = school_data_complete.loc[school_data_complete["math_score"] > 69]
math_passing_count = math_passing_df["student_name"].count()
math_passing_percent = math_passing_count/total_num_students * 100
math_passing_percent
```

```python
#Find the percentage of students with a passing (70 or greater) reading score
reading_passing_df = school_data_complete.loc[school_data_complete["reading_score"] > 69]
reading_passing_count = reading_passing_df["student_name"].count()
reading_passing_percent = reading_passing_count/total_num_students * 100
reading_passing_percent
```

```python
#Find the percentage of students who passed BOTH math and reading
both_passing_df = school_data_complete.loc[(school_data_complete["math_score"] > 69) & (school_data_complete["reading_score"] > 69)]
both_passing_count = both_passing_df["student_name"].count()
both_passing_percent = both_passing_count/total_num_students * 100
both_passing_percent
```

```python
#Create dataframe for Summary snapshot
summary_df = pd.DataFrame({"Total Number of Students": total_num_students, "Total Budget": total_budget, "Average Math Score": avg_math_score,
                           "Average Reading Score": avg_reading_score, "Percentage of Students Passing Math": math_passing_percent,
                           "Percentage of Students Passing Reading": reading_passing_percent, "Percentage of Students Passing Both": both_passing_percent}, index=[0])

#Make it look a little nicer
summary_df["Total Number of Students"] = summary_df["Total Number of Students"].map("{:,}".format)
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,}".format)
summary_df["Average Math Score"] = summary_df["Average Math Score"].map("{:,.2f}".format)
summary_df["Average Reading Score"] = summary_df["Average Reading Score"].map("{:,.2f}".format)
summary_df["Percentage of Students Passing Math"] = summary_df["Percentage of Students Passing Math"].map("{:,.2f}%".format)
summary_df["Percentage of Students Passing Reading"] = summary_df["Percentage of Students Passing Reading"].map("{:,.2f}%".format)
summary_df["Percentage of Students Passing Both"] = summary_df["Percentage of Students Passing Both"].map("{:,.2f}%".format)

#Print out the dataframe
summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Add column to original dataframe for spending per student
school_data_complete["Spending Per Student"] = school_data_complete["budget"]/school_data_complete["size"]
school_data_complete.head()
```

```python
#Create the bins and groups for School Spending
spending_bins = [0, 599, 629, 649, 700]
spending_groups = ["Less than $600", "$600 to $629", "$630 to $649", "Above $650"]
school_data_complete["Spending Range"] = pd.cut(school_data_complete["Spending Per Student"], spending_bins, labels=spending_groups, include_lowest=True)
school_data_complete2 = school_data_complete.copy()
school_data_complete2.head()
```

```python
#Group the previous dataframe by spending range
spending_groupby_df = school_data_complete2.groupby("Spending Range")

#Get a count of how many students are in each spending range
number_in_spending_range = spending_groupby_df["student_name"].count()

#Find the average passing math by spending range
avg_math_spending = spending_groupby_df["math_score"].mean()

#Find the average passing reading by spending range
avg_reading_spending = spending_groupby_df["reading_score"].mean()

#Find the percentage passing math by spending range
math_passing_df2 = school_data_complete2.loc[school_data_complete2["math_score"] > 69]
math_passing_spending_groupby_df = math_passing_df2.groupby(["Spending Range"])
math_passing_spending_count = math_passing_spending_groupby_df["student_name"].count()
math_passing_spending_percent = math_passing_spending_count/number_in_spending_range * 100

#Find the percentage passing reading by spending range
reading_passing_df2 = school_data_complete2.loc[school_data_complete2["reading_score"] > 69]
reading_passing_spending_groupby_df = reading_passing_df2.groupby(["Spending Range"])
reading_passing_spending_count = reading_passing_spending_groupby_df["student_name"].count()
reading_passing_spending_percent = reading_passing_spending_count/number_in_spending_range * 100

#Find the percentage passing both by spending range
both_passing_df2 = school_data_complete2.loc[(school_data_complete2["math_score"] > 69) & (school_data_complete2["reading_score"] > 69)]
both_passing_spending_groupby_df = both_passing_df2.groupby(["Spending Range"])
both_passing_spending_count = both_passing_spending_groupby_df["student_name"].count()
both_passing_spending_percent = both_passing_spending_count/number_in_spending_range * 100
```

```python
#Create dataframe to show data by spending range
spending_summary_df = pd.DataFrame({"Average Math Score": avg_math_spending, "Average Reading Score": avg_reading_spending,
                                    "Percent Passing Math": math_passing_spending_percent, "Percent Passing Reading": reading_passing_spending_percent,
                                    "Percent Passing Both Math and Reading": both_passing_spending_percent})
#Make it look a little nicer
spending_summary_df["Average Math Score"] = spending_summary_df["Average Math Score"].map("{:,.2f}".format)
spending_summary_df["Average Reading Score"] = spending_summary_df["Average Reading Score"].map("{:,.2f}".format)
spending_summary_df["Percent Passing Math"] = spending_summary_df["Percent Passing Math"].map("{:,.2f}%".format)
spending_summary_df["Percent Passing Reading"] = spending_summary_df["Percent Passing Reading"].map("{:,.2f}%".format)
spending_summary_df["Percent Passing Both Math and Reading"] = spending_summary_df["Percent Passing Both Math and Reading"].map("{:,.2f}%".format)

#Print out the dataframe
spending_summary_df
```
